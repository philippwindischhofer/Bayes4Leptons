```python
def index_translation_dict(df, column):
    slist = sorted(list(set(df[column])))
    print slist
    return {val: index for index, val in enumerate(slist)}
```

```python
def convert_to_matshow(df, keys, plotcol):
    translations = []
    for key in keys:
        translations.append(index_translation_dict(df, key))
                
    plot_data = np.zeros((len(translations[0]), len(translations[1]))) 
        
    for index, row in df.iterrows():
        i = translations[0][row[keys[0]]]
        j = translations[1][row[keys[1]]]

        plot_data[i,j] = row[plotcol]
       
    print translations[1]
    
    return plot_data, np.array(sorted(translations[0].keys())), np.array(sorted(translations[1].keys()))
```

```python
run_dir = "/data_CMS/cms/wind/180528_optimized_hyperparameters_selvars_leading_jets_metfix_merged/optimized/"
```

```python
# load the model configuration for this run
mconfhandler = ModelCollectionConfigFileHandler()
```

```python
mconfhandler.load_configuration(os.path.join(run_dir, "settings.conf"))
```

```python
mcolls = mconfhandler.GetModelCollection()
```

```python
df = pd.DataFrame()
for mcoll in mcolls:
    for model_name, model in mcoll.model_dict.iteritems():
        row_dict = dict(model.hyperparameters)
        row_dict["number_layers"] += 1
        row_dict["model_name"] = model_name
        row_dict["count"] = 1
        df = df.append(row_dict, ignore_index = True)
```

```python
df
```

```python
df_reduced = df[["count", "number_layers", "number_neurons"]]
```

```python
df_counts = df_reduced.groupby(["number_layers", "number_neurons"], group_keys = False).sum()
```

```python
df_final = pd.DataFrame()
```

```python
for row in df_counts.iterrows():
    row_dict = {"number_layers": row[0][0],
               "number_neurons": row[0][1],
               "counts": row[1][0]}
    df_final = df_final.append(row_dict, ignore_index = True)
```

```python
df_final
```

```python
data, x_label, y_label = convert_to_matshow(df_final, ["number_layers", "number_neurons"], "counts")
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
cax = ax.matshow(data.transpose(), interpolation = 'nearest', cmap = 'Blues', vmin = np.min(data), vmax = np.max(data))
ax.set_xticklabels(np.concatenate([[''], x_label.astype(int)]))
ax.set_yticklabels(np.concatenate([[''], y_label.astype(int)]))
ax.xaxis.set_label_position("bottom")
ax.xaxis.tick_bottom()
ax.set_xlabel("hidden layers")
ax.set_ylabel("neurons / hidden layer")

for cur in np.ndenumerate(data):
    coords = cur[0]
    val = cur[1]
    
    if val > 3:
        color = 'white'
    else:
        color = 'black'
        
    ax.text(coords[0], coords[1], "{:4.0f}".format(val), va = 'center', ha = 'center', color = color, weight = 'bold')
```

```python
x_ticklabels = x_label.astype(int).astype(str)
y_ticklabels = y_label.astype(int).astype(str)
```

```python
def root_matshow(data, x_ticklabels, y_ticklabels, x_axislabel, y_axislabel, show_values, outfile):
    c = ROOT.TCanvas("canv", "canv", 800, 800)
    
    rows = np.shape(data)[0]
    cols = np.shape(data)[1]
    
    num_colors = 3;
                        
    reds = np.array([0.9686274528503418,0.41708574, 0.031372550874948502])
    greens = np.array([0.9843137264251709, 0.68063055, 0.18823529779911041])
    blues = np.array([1.00, 0.83823146, 0.41960784792900085])
    stops = np.array([0.0, 0.5, 1.0])
    
    num_internal_colors = 150;
    ROOT.TColor.CreateGradientColorTable(num_colors, stops, reds, greens, blues, num_internal_colors);
    
    hist = ROOT.TH2F("matshow", "", rows, -0.5, rows - 0.5, cols, -0.5, cols - 0.5)
    ROOT.gStyle.SetOptStat(0)

    for row in range(rows):
        for col in range(cols):
            hist.Fill(row, col, data[row, col])
            
    # do the axis tick labels and axis labels
    hist.GetXaxis().SetLabelOffset(100)
    hist.GetYaxis().SetLabelOffset(100)
    
    c.cd()
    c.SetLeftMargin(0.15)
    c.SetBottomMargin(0.15)
    hist.Draw("col")
    
    t = ROOT.TLatex()
    t.SetTextAlign(22)
    t.SetTextSize(0.025)
        
    y = ROOT.gPad.GetUymin() - 0.7 * hist.GetYaxis().GetBinWidth(1)
    for i, label in enumerate(x_ticklabels):
        x = hist.GetXaxis().GetBinCenter(i + 1)
        t.DrawLatex(x, y, label)
        
    t.SetTextSize(0.03)
    t.DrawLatex(0.5 * hist.GetXaxis().GetBinCenter(rows), y - 0.2 * hist.GetYaxis().GetBinWidth(1), x_axislabel)
    t.SetTextSize(0.025)

    x = ROOT.gPad.GetUxmin() - 0.7 * hist.GetXaxis().GetBinWidth(1)
    for i, label in enumerate(y_ticklabels):
        y = hist.GetYaxis().GetBinCenter(i + 1)
        t.DrawLatex(x, y, label)
        
    t.SetTextSize(0.03)
    t.SetTextAngle(90)
    t.SetTextAlign(22)
    t.DrawLatex(x - 0.3 * hist.GetXaxis().GetBinWidth(1), 0.5 * hist.GetYaxis().GetBinCenter(rows), y_axislabel)
    t.SetTextSize(0.025)
    
    t.SetTextAngle(0)
    if show_values:
        for row in range(rows):
            for col in range(cols):
                x = hist.GetXaxis().GetBinCenter(col + 1)
                y = hist.GetYaxis().GetBinCenter(row + 1)
                if hist.GetBinContent(col + 1, row + 1) > 2:
                    t.SetTextColor(ROOT.kWhite)
                else:
                    t.SetTextColor(ROOT.kBlack)
                t.DrawLatex(x, y, str(int(hist.GetBinContent(col + 1, row + 1))))
                
    ROOT.gPad.RedrawAxis()
    c.SetTicks(1,1)
    c.Update()
        
    c.SaveAs(outfile)
```

```python
root_matshow(np.fliplr(data), x_ticklabels, np.flipud(y_ticklabels), "hidden layers", "hidden neurons / layer", True, "/data_CMS/cms/wind/HyperparameterPlots/model_histogram_rootstyle.pdf")
```
